```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
hpa_BA = pd.read_csv('./hpa_bileacids.tsv', sep='\t', index_col=0)
hpa_BA
```

```python
set(hpa_BA.index)
```

```python
hpa_BA.columns
```

```python
hpa_loc_BA = hpa_BA[['RNA tissue specific nTPM']].copy()
hpa_loc_BA.head()
```

```python
print(list(hpa_loc_BA['RNA tissue specific nTPM'].unique()))
```

```python
organ_list = ['liver', 'gallbladder', 'pancreas', 'intestine', 'stomach', 
              'kidney', 'adrenal', 'spleen', 
              'ovary', 'endometrium', 'cervix', 'breast', 'epididymis', 'testis',
              'heart', 'lung',
              'adipose', 'lymphoid', 'skeletal', 'smooth', 'bone', 'skin', 
              'parathyroid', 'thyroid', 'esophagus', 'tongue', 
              'retina', 'pituitary', 'brain']
```

```python
for organ in organ_list:
    hpa_loc_BA[organ] = hpa_loc_BA['RNA tissue specific nTPM'].str.contains(organ)
    hpa_loc_BA[organ].replace({True: 1, False: 0}, inplace=True)
```

```python
hpa_loc_BA
```

```python
hpa_loc_BA.shape
```

```python
hpa_loc_BA.dropna(inplace=True)
```

```python
hpa_loc_BA.shape
```

```python
hpa_loc_BA.drop(columns='RNA tissue specific nTPM', inplace=True)
```

```python
hpa_loc_BA.head()
```

## Subset of Known Genes

```python
hpa_BA.index
```

```python
known_genes = ['NR1H4', 'BAAT', 'GPBAR1', 'VDR', 
               'NR1I1', 'NR1I2', 'NR1I3',
               'CYP7A1', 'CYP8B1', 
               'SLC10A1', 'SLC10A2','SLC51A', 'SLC51A', 'SLC27A2', 'SLC27A5',
               'AKR1D1', 'SULT2A1']
```

```python
known_BA = hpa_BA[hpa_BA.index.isin(known_genes)]
known_BA.shape
```

```python
known_BA.head()
```

```python
known_loc_BA = known_BA[['RNA tissue specific nTPM']].copy()
known_loc_BA.head()
```

```python
for organ in organ_list:
    known_loc_BA[organ] = known_loc_BA['RNA tissue specific nTPM'].str.contains(organ)
    known_loc_BA[organ].replace({True: 1, False: 0}, inplace=True)
```

```python
known_loc_BA.drop(columns='RNA tissue specific nTPM', inplace=True)
```

```python
known_loc_BA.dropna(inplace=True)
```

```python
known_loc_BA2 = known_loc_BA.loc[:, (known_loc_BA != 0).any(axis=0)]
known_loc_BA2
```

```python
sns.heatmap(data=known_loc_BA2.reindex(known_genes), cmap='Blues', 
              xticklabels=True, yticklabels=True)
#plt.savefig('./BA_bileacid_knowngene_loc_heatmap.svg', dpi=150, bbox_inches='tight')
```

```python
sns.clustermap(data=known_loc_BA2, cmap='Blues', col_cluster=False,
              xticklabels=True, yticklabels=True)
#plt.savefig('./BA_bileacid_knowngene_loc_heatmap.svg', dpi=150, bbox_inches='tight')
```
